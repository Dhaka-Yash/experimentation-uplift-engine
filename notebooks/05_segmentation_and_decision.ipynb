```python
avg_order_value = 100

if "df" not in globals():
    import pandas as pd
    df = pd.read_csv("D:/My Projects/experiment-uplifting engine/data/experiment_simulated.csv")

treatment_rate = df.loc[df["treatment"] == 1, "converted"].mean()
control_rate = df.loc[df["treatment"] == 0, "converted"].mean()
absolute_lift = treatment_rate - control_rate

incremental_revenue = absolute_lift * len(df) * avg_order_value
print("Absolute lift:", round(absolute_lift, 4))
print("Estimated incremental revenue:", round(incremental_revenue, 2))
```

Based on statistically significant uplift of 0.015%
we recommend rolling out to new users and mobile users only.
