```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest

df = pd.read_csv("D:/My Projects/experiment-uplifting engine/data/experiment_simulated.csv")
```

```python
summary = df.groupby('treatment')['converted'].mean()
control_cr = summary[0]
treatment_cr = summary[1]

absolute_lift = treatment_cr - control_cr
relative_lift = absolute_lift / control_cr

print("Control CR:", control_cr)
print("Treatment CR:", treatment_cr)
print("Absolute Lift:", absolute_lift)
print("Relative Lift:", relative_lift)
```

```python
counts = df.groupby('treatment')['converted'].sum()
nobs = df.groupby('treatment')['converted'].count()

z_stat, p_value = proportions_ztest(counts, nobs)

print("P-value:", p_value)
```

```python
import statsmodels.api as sm

ci_low, ci_upp = sm.stats.proportion_confint(
    counts[1], nobs[1], alpha=0.05, method='normal'
)

print("Treatment CI:", ci_low, ci_upp)
```
