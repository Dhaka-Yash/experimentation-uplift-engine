```python
import pandas as pd
from statsmodels.stats.power import NormalIndPower
from statsmodels.stats.proportion import proportion_effectsize
```

```python
baseline = 0.08
expected = 0.10
alpha = 0.05
power = 0.8
```

```python
effect_size = proportion_effectsize(baseline, expected)

analysis = NormalIndPower()
required_n = analysis.solve_power(effect_size=effect_size,
                                  power=power,
                                  alpha=alpha)

print("Required sample per group:", round(required_n))
```
